```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Ler o arquivo
df = pd.read_csv("dataset1.csv", sep=";")

#Mostra tudo
pd.set_option('display.max_rows', None)

#Testes
df['US 2024 Deficit'] = df['US 2024 Deficit'].str.replace(",", "").astype(float)

print(df[df['US 2024 Deficit'] < -30])
```

```python
#Países com deficit positivo
country_value = df[df["US 2024 Deficit"] > 0]["Country"]
print(country_value)
```

```python
#Gráfico

top10 = df.sort_values('US Deficit 2024', ascending=False).head(10)
plt.Figure(figsize=(10,6))
plt.xlabel("Country")
plt.ylabel("US Deficit 2024")
plt.bar()
plt.title = "Top 10 Deficit nos EUA 2024"
plt.tight_layout()
plt.show()
```
